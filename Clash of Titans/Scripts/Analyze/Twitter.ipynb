FEELINGS

```python
import pandas as pd
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import nltk

# Baixar o lexicon do VADER
#nltk.download('vader_lexicon')

# Caminhos para os arquivos Excel
xgstwitter_path = r'C:\Users\Shelton\Documents\VSCode\Python\Clash of Titans\Data\Clean\Main\XGS Twitter.xlsx'
soptwitter_path = r'C:\Users\Shelton\Documents\VSCode\Python\Clash of Titans\Data\Clean\Main\SoP Twitter.xlsx'

# Carregar os arquivos Excel em DataFrames
xgstwitter_df = pd.read_excel(xgstwitter_path)
soptwitter_df = pd.read_excel(soptwitter_path)

# Inicializar o analisador de sentimentos do VADER
sia = SentimentIntensityAnalyzer()

# Função para calcular o sentimento e aplicar rótulos
def get_sentiment(text):
    score = sia.polarity_scores(text)
    if score['compound'] > 0.05:
        return 'Positive'
    elif score['compound'] < -0.05:
        return 'Negative'
    else:
        return 'Neutral'

# Garantir que os comentários são strings e lidar com valores ausentes
xgstwitter_df['Comments'] = xgstwitter_df['Comments'].astype(str).fillna('')
soptwitter_df['Comments'] = soptwitter_df['Comments'].astype(str).fillna('')

# Aplicar a função aos comentários nos DataFrames
xgstwitter_df['Sentiment'] = xgstwitter_df['Comments'].apply(get_sentiment)
soptwitter_df['Sentiment'] = soptwitter_df['Comments'].apply(get_sentiment)

# Exibir os DataFrames com a nova coluna de sentimento
print(xgstwitter_df.head())
print(soptwitter_df.head())

# Se desejar, salvar os DataFrames de volta nos arquivos Excel
xgstwitter_df.to_excel(r'C:\Users\Shelton\Documents\VSCode\Python\Clash of Titans\Data\Clean\Main\Feelings\XGS Twitter Feelings.xlsx', index=False)
soptwitter_df.to_excel(r'C:\Users\Shelton\Documents\VSCode\Python\Clash of Titans\Data\Clean\Main\Feelings\SoP Twitter Feelings.xlsx', index=False)
```

```python
import pandas as pd
import text2emotion as te
from tqdm import tqdm

# Caminhos para os arquivos Excel
xgstwitter_path = r'C:\Users\Shelton\Documents\VSCode\Python\Clash of Titans\Data\Clean\Main\Feelings\XGS Twitter Feelings.xlsx'
soptwitter_path = r'C:\Users\Shelton\Documents\VSCode\Python\Clash of Titans\Data\Clean\Main\Feelings\SoP Twitter Feelings.xlsx'

# Carregar os arquivos Excel em DataFrames
xgstwitter_df = pd.read_excel(xgstwitter_path)
soptwitter_df = pd.read_excel(soptwitter_path)

# Função para calcular a emoção dominante
def get_dominant_emotion(text):
    try:
        emotions = te.get_emotion(text)
        # Retornar a emoção com o maior valor
        dominant_emotion = max(emotions, key=emotions.get)
        return dominant_emotion
    except Exception as e:
        print(f"Error processing text: {text}")
        print(f"Exception: {e}")
        return None

# Usar tqdm para acompanhar o progresso
tqdm.pandas()

# Aplicar a função aos comentários nos DataFrames com progresso
xgstwitter_df['Emotion'] = xgstwitter_df['Comments'].progress_apply(get_dominant_emotion)
soptwitter_df['Emotion'] = soptwitter_df['Comments'].progress_apply(get_dominant_emotion)

# Exibir os DataFrames com a nova coluna de sentimento dominante
print(xgstwitter_df.head())
print(soptwitter_df.head())

# Se desejar, salvar os DataFrames de volta nos arquivos Excel
xgstwitter_df.to_excel(r'C:\Users\Shelton\Documents\VSCode\Python\Clash of Titans\Data\Clean\Main\Feelings\XGS Twitter Emotions.xlsx', index=False)
soptwitter_df.to_excel(r'C:\Users\Shelton\Documents\VSCode\Python\Clash of Titans\Data\Clean\Main\Feelings\SoP Twitter Emotions.xlsx', index=False)
```
